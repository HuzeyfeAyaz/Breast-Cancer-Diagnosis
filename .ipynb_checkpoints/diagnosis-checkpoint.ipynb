# Breast Cancer Data-Set

In this ipython file we tried mainly 2 approaches namely, 1.1 and 1.2. Since this is the our first data-science project may includes some mistakes.

Approach 1.1: Creating a multi-class classifier to predict if a given individual is healthy, ER, PR, or Her-2 neu. For this approach we eliminated the multilabel cases which is each patient has more than one subtype.

Approach 1.2: Instead of multi-class classifier for this case we created different binary classifier to evaluate them separately.

### Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
import matplotlib.patches as mpatches
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from itertools import product
```

```python
import warnings
warnings.simplefilter('ignore')
```

### Approach 1.1

```python
df = pd.read_excel('COH plasma data without multilabel.xlsx', sheet_name='Sheet1')
df = df.dropna(axis=0, how='all', inplace=False)
```

```python
types = np.array(df.iloc[1:4,8:].fillna(-1)).T
metabolites = np.array(df.iloc[16:,8:]).T
```

```python
# Finding roc_curve and Auc score for each classifier
def find_roccurve(classifier,X_test,y_test):
    y_pred = classifier.predict(X_test)
    y_prob = classifier.predict_proba(X_test)[:,1]
    auc = roc_auc_score(y_test, y_prob)  
    # print('AUC: %.2f' % auc)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    return (auc, fpr, tpr, thresholds)
```

```python
target = []
for i in types:
    for a in range(len(i)):
        if i[a] == 1:
            target.append(a+1)
        if i[a] == -1:
            target.append(0)
            break
target = np.ravel(target)
```

```python
# Standart Scaler
sc = StandardScaler()
metabolites_std = sc.fit_transform(metabolites)

# dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(metabolites_std, target, test_size = 0.3, random_state = 0)
```

```python
scores = []
# SVM (Support Vector Machine)
svm = SVC(C=10.0**0, kernel='sigmoid', random_state=0, gamma=10**-2, probability=True) # 826
svm_score = round(sum(cross_val_score(svm, metabolites_std, y=target, scoring='f1_micro', cv=10))*10,2)
scores.append(svm_score)

# Logistic Regression
lr = LogisticRegression(penalty='l1', random_state=0, C=10.0**-1, max_iter=990) # 0.8247
lr_score = round(sum(cross_val_score(lr, metabolites_std, y=target, scoring='f1_micro', cv=10))*10,2)
scores.append(lr_score)

# Random Forest
rf = RandomForestClassifier(n_estimators=40) # 0.85
rf_score = round(sum(cross_val_score(rf, metabolites_std, y=target, scoring='f1_micro', cv=10))*10,2)
scores.append(rf_score)

# Decision Tree
tree = DecisionTreeClassifier(criterion='entropy', max_depth=1, random_state=0) # 84.07
tree_score = round(sum(cross_val_score(tree, metabolites_std, y=target, scoring='f1_micro', cv=10))*10,2)
scores.append(tree_score)

# knn
knn = KNeighborsClassifier(n_neighbors=7, p=2, metric='minkowski') # 77.96
knn_score = round(sum(cross_val_score(knn, metabolites_std, y=target, scoring='f1_micro', cv=10))*10,2)
scores.append(knn_score)
```

```python
y_pos = np.arange(len(scores))
plt.figure(figsize=(12,5))
x1 = plt.subplot(121)
classifiers = ['SVM','LR','RF','DT', 'KNN']
clr = ('orange','blue', 'green', 'purple', 'red', 'purple')

x1.bar(y_pos, scores, align='center', alpha=1.0, color=clr)
locs, labels = plt.xticks(y_pos,classifiers)
x1.set_yticks([i for i in range(0,110,10)])
legends = []
for i in range(len(classifiers)):
    legends.append(mpatches.Patch(color=clr[i], label='{}'.format(scores[i])))
x1.legend(handles=legends,loc='lower left')
x1.set_xlabel('Classifiers')
x1.set_ylabel('F-1 Scores')
x1.title.set_text('10-K Scores for approach 1.1 (Multi-Class Classification)')

# Roc curve
x2 = plt.subplot(122)
# Binarize the output
binarized_target = label_binarize(target, classes=[0, 1, 2, 3])
n_classes = binarized_target.shape[1]
X_train, X_test, y_train, y_test = train_test_split(metabolites_std, binarized_target, test_size = 0.3, random_state = 0)

# Learn to predict each class against the other
classifier = OneVsRestClassifier(svm)
y_score = classifier.fit(X_train, y_train).decision_function(X_test)

# Compute ROC curve and ROC area for each class
keep_roc = dict()
for i in range(n_classes):
    fpr_, tpr_, threshold = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc = roc_auc_score(y_test[:, i], y_score[:, i])
    keep_roc[i] = (roc_auc, fpr_, tpr_, threshold)

classes = ['healthy_Patient', 'Pn', 'Er', 'Her_2 Neu']
colors = ['aqua', 'darkorange', 'cornflowerblue','red']
for i, color in zip(range(n_classes), colors):
    x2.plot(keep_roc[i][1], keep_roc[i][2], color=color,
             label='{0} (area = {1:0.2f})'.format(classes[i], keep_roc[i][0]))

x2.plot([0, 1], [0, 1], 'k--')
x2.set_xlabel('False Positive Rate')
x2.set_ylabel('True Positive Rate')
x2.title.set_text('Roc_curve for multi-class classification by SVM')
x2.legend(loc="lower right")
plt.savefig('approach_1.1.png', dpi=300)
plt.show()
```

### Approach 1.2

```python
df = pd.read_excel('COH plasma data with multilabel.xlsx', sheet_name='Sheet1')
df = df.dropna(axis=0, how='all', inplace=False)
```

```python
# taking each label from dataset
patients = np.array(df.iloc[15,8:]).astype('int')
all_metabolites = np.array(df.iloc[16:,8:]).T
subtypes = df.iloc[:,8:]
subtypes = subtypes.dropna(axis=1, thresh=(subtypes.shape[0]-7))
subtypes_metabolites = np.array(subtypes.iloc[16:,:]).T
pr = np.array(subtypes.iloc[1,:]).astype('int')
er = np.array(subtypes.iloc[2,:]).astype('int')
her_neu = np.array(subtypes.iloc[3,:]).astype('int')
sc = StandardScaler()
sc1 = StandardScaler()
metabolites_std = sc.fit_transform(all_metabolites)
subtypes_std = sc1.fit_transform(subtypes_metabolites)
```

```python
# Train and Test set
X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer = train_test_split(all_metabolites, patients, test_size = 0.3, random_state = 1)
X_train_pr, X_test_pr, y_train_pr, y_test_pr = train_test_split(subtypes_metabolites, pr, test_size = 0.3, random_state = 1)
X_train_er, X_test_er, y_train_er, y_test_er = train_test_split(subtypes_metabolites, er, test_size = 0.3, random_state = 1)
X_train_her, X_test_her, y_train_her, y_test_her = train_test_split(subtypes_metabolites, her_neu, test_size = 0.3, random_state = 1)
```

### Dimensionality reduction tecniques

```python
def find_val_score(classifier, X, y, name='f1_micro'):
    scores = cross_val_score(estimator=classifier,
                         X=X,
                         y=y,
                         cv=5,
                         scoring='f1_micro')
    return ("%s: %0.2f (+/- %0.2f)" % (name, scores.mean(), scores.std()), scores.mean(), scores.std())

def find_best(classifier, X, y, param_grid):
    gs = GridSearchCV(estimator=classifier, 
                      param_grid=param_grid, 
                      scoring='f1_micro',
                      cv=10,
                      n_jobs=-1)
    gs = gs.fit(X, y)
    return (gs.best_score_,gs.best_params_)
```

```python
max_depth = [None,4,5]
neighbors = [i for i in range(1,11)]
param_range = [0.01, 0.1, 1.0, 10.0, 100.0]

# Param Grids for lda
param_lda_lr = [{'clf__C': param_range,
                  'clf__penalty': ['l1','l2']}]

param_lda_svm = [{'clf__C': param_range,
                  'clf__kernel': ['sigmoid','rbf','linear']}]

param_lda_dt = [{'clf__max_depth': max_depth,
                  'clf__criterion': ['gini','entropy']}]

param_lda_knn = [{'clf__n_neighbors': neighbors}]


# Param Grids for ensembles
param_grid1 = [{'clf__LogisticRegression__C': param_range, 
               'clf__LogisticRegression__penalty': ['l1','l2'],
               'clf__DecisionTree__max_depth': max_depth,
               'clf__DecisionTree__criterion': ['gini','entropy'],
               'clf__KNN__n_neighbors': neighbors
               }]

param_grid2 = [{'clf__base_estimator__max_depth': max_depth,
                'clf__base_estimator__criterion': ['gini','entropy'],
                }]

param_grid_comb = [param_grid1, param_grid2, param_grid2]
param_grid_comb2 = [param_lda_lr, param_lda_svm, param_lda_dt, param_lda_knn]
```

```python
def find_pca_f1(X, y, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe_pca = Pipeline([('scl', StandardScaler()),
                            ('reducer', PCA(0.95)),
                            ('clf', clf)])
        pipe_pca1 = Pipeline([('scl', StandardScaler()),
                            ('reducer', PCA(0.95)),
                            ('clf', clf)])

        grid = find_best(pipe_pca, X, y, param)
        pipe_pca1.set_params(**grid[1])
        params = grid[1]
        
        scores = cross_val_score(estimator=pipe_pca1, X=X, y=y, cv=10, scoring='f1_micro')
        clf_scores[name] = ("%s: %0.2f (+/- %0.2f)" % (name, scores.mean(), scores.std()), scores.mean(), scores.std(), params)
    return clf_scores

def find_lda_f1(X, y, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe_pca = Pipeline([('scl', StandardScaler()),
                            ('reducer', LDA()),
                            ('clf', clf)])
        pipe_pca1 = Pipeline([('scl', StandardScaler()),
                            ('reducer', LDA()),
                            ('clf', clf)])

        grid = find_best(pipe_pca, X, y, param)
        pipe_pca1.set_params(**grid[1])
        params = grid[1]
        
        scores = cross_val_score(estimator=pipe_pca1, X=X, y=y, cv=10, scoring='f1_micro')
        clf_scores[name] = ("%s: %0.2f (+/- %0.2f)" % (name, scores.mean(), scores.std()), scores.mean(), scores.std(), params)
    return clf_scores

def find_f1(X, y, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe_pca = Pipeline([('scl', StandardScaler()),
                            ('clf', clf)])
        pipe_pca1 = Pipeline([('scl', StandardScaler()),
                            ('clf', clf)])

        grid = find_best(pipe_pca, X, y, param)
        pipe_pca1.set_params(**grid[1])
        params = grid[1]
        
        scores = cross_val_score(estimator=pipe_pca1, X=X, y=y, cv=10, scoring='f1_micro')
        clf_scores[name] = ("%s: %0.2f (+/- %0.2f)" % (name, scores.mean(), scores.std()), scores.mean(), scores.std(), params)
    return clf_scores
```

```python
def find_pca_roc(X_train, X_test, y_train, y_test, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(probability=True), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe_pca = Pipeline([('scl', StandardScaler()),
                            ('reducer', PCA()),
                            ('clf', clf)])
        pipe_pca1 = Pipeline([('scl', StandardScaler()),
                            ('reducer', PCA()),
                            ('clf', clf)])

        grid = find_best(pipe_pca, X_train, y_train, param)
        pipe_pca1.set_params(**grid[1])
        pipe_pca1.fit(X_train, y_train)
        clf_scores[name] = find_roccurve(pipe_pca1, X_test, y_test)
    return clf_scores

def find_lda_roc(X_train, X_test, y_train, y_test, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(probability=True), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe_lda = Pipeline([('scl', StandardScaler()),
                            ('reducer', LDA()),
                            ('clf', clf)])
        pipe_lda1 = Pipeline([('scl', StandardScaler()),
                            ('reducer', LDA()),
                            ('clf', clf)])

        grid = find_best(pipe_lda, X_train, y_train, param)
        pipe_lda1.set_params(**grid[1])
        pipe_lda1.fit(X_train, y_train)        
        clf_scores[name] = find_roccurve(pipe_lda1, X_test, y_test)
    return clf_scores

def find_roc(X_train, X_test, y_train, y_test, param_grids):
    clf_scores = {}
    classifiers = [LogisticRegression(random_state=0), SVC(probability=True), DecisionTreeClassifier(), KNeighborsClassifier()]
    labels = ['LR', 'SVM', 'DT', 'KNN']

    for clf, param, name in zip(classifiers, param_grids, labels):
        pipe = Pipeline([('scl', StandardScaler()),
                            ('clf', clf)])
        pipe1 = Pipeline([('scl', StandardScaler()),
                            ('clf', clf)])

        grid = find_best(pipe, X_train, y_train, param)
        pipe1.set_params(**grid[1])
        pipe1.fit(X_train, y_train)
        clf_scores[name] = find_roccurve(pipe1, X_test, y_test)
    return clf_scores
```

## F1 Score and Roc Curve Calculation

```python
# Roc curves with PCA
patient_pca_roc = find_pca_roc(X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer, param_grid_comb2)
pr_pca_roc = find_pca_roc(X_train_pr, X_test_pr, y_train_pr, y_test_pr, param_grid_comb2)
er_pca_roc = find_pca_roc(X_train_er, X_test_er, y_train_er, y_test_er, param_grid_comb2)
her_pca_roc = find_pca_roc(X_train_her, X_test_her, y_train_her, y_test_her, param_grid_comb2)

# Roc curves with LDA
patient_lda_roc = find_lda_roc(X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer, param_grid_comb2)
pr_lda_roc = find_lda_roc(X_train_pr, X_test_pr, y_train_pr, y_test_pr, param_grid_comb2)
er_lda_roc = find_lda_roc(X_train_er, X_test_er, y_train_er, y_test_er, param_grid_comb2)
her_lda_roc = find_lda_roc(X_train_her, X_test_her, y_train_her, y_test_her, param_grid_comb2)

# Roc curves
patient_roc = find_roc(X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer, param_grid_comb2)
pr_roc = find_roc(X_train_pr, X_test_pr, y_train_pr, y_test_pr, param_grid_comb2)
er_roc = find_roc(X_train_er, X_test_er, y_train_er, y_test_er, param_grid_comb2)
her_roc = find_roc(X_train_her, X_test_her, y_train_her, y_test_her, param_grid_comb2)
```

```python
# F1 scores with PCA
patient_pca_f1 = find_pca_f1(all_metabolites, patients, param_grid_comb2)
pr_pca_f1 = find_pca_f1(subtypes_metabolites, pr, param_grid_comb2)
er_pca_f1 = find_pca_f1(subtypes_metabolites, er, param_grid_comb2)
her_pca_f1 = find_pca_f1(subtypes_metabolites, her_neu, param_grid_comb2)

# F1 scores with LDA
patient_lda_f1 = find_lda_f1(all_metabolites, patients, param_grid_comb2)
pr_lda_f1 = find_lda_f1(subtypes_metabolites, pr, param_grid_comb2)
er_lda_f1 = find_lda_f1(subtypes_metabolites, er, param_grid_comb2)
her_lda_f1 = find_lda_f1(subtypes_metabolites, her_neu, param_grid_comb2)

# F1 scores
patient_f1 = find_f1(all_metabolites, patients, param_grid_comb2)
pr_f1 = find_f1(subtypes_metabolites, pr, param_grid_comb2)
er_f1 = find_f1(subtypes_metabolites, er, param_grid_comb2)
her_f1 = find_f1(subtypes_metabolites, her_neu, param_grid_comb2)
```

## Learning and Validation Curves

```python
from sklearn.model_selection import learning_curve
from sklearn.model_selection import validation_curve
```

```python
def plot_learning_curve(estimator, X, y, title, plot_):
    train_sizes, train_scores, test_scores =\
                learning_curve(estimator=estimator,
                               X=X,
                               y=y,
                               train_sizes=np.linspace(0.1, 1.0, 10),
                               cv=10,
                               n_jobs=1)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plot_.plot(train_sizes, train_mean,
             color='blue', marker='o',
             markersize=5, label='training accuracy')

    plot_.fill_between(train_sizes,
                     train_mean + train_std,
                     train_mean - train_std,
                     alpha=0.15, color='blue')

    plot_.plot(train_sizes, test_mean,
             color='green', linestyle='--',
             marker='s', markersize=5,
             label='validation accuracy')

    plot_.fill_between(train_sizes,
                     test_mean + test_std,
                     test_mean - test_std,
                     alpha=0.15, color='green')

    plot_.grid()
    plot_.set_xlabel('Number of training samples')
    plot_.set_ylabel('Accuracy')
    plot_.legend(loc='lower right')
    plot_.set_ylim([0, 1.01])
#     plot_.tight_layout()
    plot_.title.set_text('Learning Curve For {} by LR'.format(title))
#     plt.savefig('learning_curve_1_2.png', dpi=300)
```

```python
def randomize(X, Y):
     permutation = np.random.permutation(Y.shape[0])
     X2 = X[permutation,:]
     Y2 = Y[permutation]
     return X2, Y2
```

```python
# New X and Y values
X1, Y1 = randomize(metabolites_std, patients)
X2, Y2 = randomize(subtypes_std, pr)
X3, Y3 = randomize(subtypes_std, er)
X4, Y4 = randomize(subtypes_std, her_neu)
```

## Learning Curves

```python
cancer_healthy_lr = LogisticRegression() #95.38
pr_lr = LogisticRegression(penalty='l1', C=10.0**-1,random_state=0,max_iter=990) # 72.83
er_lr = LogisticRegression(penalty='l1', C=10.0**-1,random_state=0) # 60.31
her_neu_lr = LogisticRegression(penalty='l2', C=10.0**0,random_state=0) # 58.84

# Plots
fig = plt.figure(figsize=(12,10))

x1 = plt.subplot(221)
plot_learning_curve(cancer_healthy_lr, X1, Y1, 'Patient-Healthy', x1)
x2 = plt.subplot(222)
plot_learning_curve(pr_lr, X2, Y2, 'PR', x2)
x3 = plt.subplot(223)
plot_learning_curve(er_lr, X3, Y3, 'ER', x3)
x4 = plt.subplot(224)
plot_learning_curve(her_neu_lr, X4, Y4, 'Her_2 Neu', x4)
fig.suptitle('Learning Curve Without Dimensionally Reduction',y=0.95, fontsize=16)
plt.savefig('LearningCurve_1_2_without_DR', dpi=300)
plt.show()
```

## Validation Curves

```python
def plot_validation_curve(estimator, X, y, title, plot_):
    param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
    train_scores, test_scores = validation_curve(
                    estimator=estimator, 
                    X=X, 
                    y=y, 
                    param_name='C', 
                    param_range=param_range,
                    cv=10)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plot_.plot(param_range, train_mean, 
             color='blue', marker='o', 
             markersize=5, label='training accuracy')

    plot_.fill_between(param_range, train_mean + train_std,
                     train_mean - train_std, alpha=0.15,
                     color='blue')

    plot_.plot(param_range, test_mean, 
             color='green', linestyle='--', 
             marker='s', markersize=5, 
             label='validation accuracy')

    plot_.fill_between(param_range, 
                     test_mean + test_std,
                     test_mean - test_std, 
                     alpha=0.15, color='green')

    plot_.grid()
    plot_.set_xscale('log')
    plot_.legend(loc='lower right')
    plot_.set_xlabel('Parameter C')
    plot_.set_ylabel('Accuracy')
    plot_.set_ylim([0, 1.01])
    plot_.title.set_text('Validation Curve For {} by LR'.format(title))
    # plt.tight_layout()
    # plt.show()
```

```python
lr = LogisticRegression(penalty='l2', random_state=0)
# Plots
fig = plt.figure(figsize=(12,10))

x1 = plt.subplot(221)
plot_validation_curve(lr, metabolites_std, patients, 'Patient-Healthy', x1)
x2 = plt.subplot(222)
plot_validation_curve(lr, subtypes_std, pr, 'PR', x2)
x3 = plt.subplot(223)
plot_validation_curve(lr, subtypes_std, er, 'ER', x3)
x4 = plt.subplot(224)
plot_validation_curve(lr, subtypes_std, her_neu, 'Her_2 Neu', x4)
fig.suptitle('Validation Curve Without Dimensionally Reduction',y=0.95, fontsize=16)
plt.savefig('Validation_curve_1_2_without_DR', dpi=300)
plt.show()
```

## Confusion Matrix

```python
def plot_confusion(confmat, ax):
    ax.matshow(confmat, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(confmat.shape[0]):
        for j in range(confmat.shape[1]):
            ax.text(x=j, y=i, s=confmat[i, j], va='center', ha='center')

    plt.xlabel('predicted label')
    plt.ylabel('true label')
```

```python
fig, ax = plt.subplots(figsize=(6, 7))
x1 = plt.subplot(221)
cancer_healthy_lr.fit(X_train_cancer, y_train_cancer)
y_pred = cancer_healthy_lr.predict(X_test_cancer)
confmat1 = confusion_matrix(y_true=y_test_cancer, y_pred=y_pred)
plot_confusion(confmat1,x1)
x1.title.set_text('Patient-Healthy')

x2 = plt.subplot(222)
pr_lr.fit(X_train_pr, y_train_pr)
y_pred = pr_lr.predict(X_test_pr)
confmat2 = confusion_matrix(y_true=y_test_pr, y_pred=y_pred)
plot_confusion(confmat2,x2)
x2.title.set_text('PR')

x3 = plt.subplot(223)
er_lr.fit(X_train_er, y_train_er)
y_pred = er_lr.predict(X_test_er)
confmat3 = confusion_matrix(y_true=y_test_er, y_pred=y_pred)
plot_confusion(confmat3,x3)
x3.title.set_text('ER')

x4 = plt.subplot(224)
her_neu_lr.fit(X_train_her, y_train_her)
y_pred = her_neu_lr.predict(X_test_her)
confmat4 = confusion_matrix(y_true=y_test_her, y_pred=y_pred)
plot_confusion(confmat4,x4)
x4.title.set_text('Her_2 Neu')

plt.tight_layout()
fig.suptitle('Confusion Matrix without Dimensionally Reduction by LR',y=0.99, fontsize=14)
plt.savefig('Confusion_matrix_1_2_without_DR', dpi=300)
plt.show()
```
